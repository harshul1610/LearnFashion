### Data Preprocessing 1 - Jabong

This code gets all the images from jabong and put them in jabong folder. It also prepares the all_jabong.csv for the tefla ready format. all_jabong is present in jabong folder only. 

```python
import os
import glob
import random
import json
import pandas as pd
import urllib
import sys
```

```python
home_dir = '../Master-Scrapper/json/jabong/'
csv_path = 'json/'
images_path = 'jabong/'

try:
    os.mkdir('jabong')
except:
    pass

all_files = glob.glob(home_dir+ '*.json')

i=0
jabong_dataframe = pd.DataFrame()
jabong_dataframe_imageurl = []
jabong_dataframe_category = []
jabong_dataframe_image = []
jabong_dataframe_desc = []
jabong_dataframe_name = []
jabong_dataframe_price = []


for file_ in all_files:
    with open(file_,'r') as f:
        data = f.read()
        json_data = json.loads(data)
        product_category = json_data['category']
        products = json_data['Products']
        for product in products:
            try:
                print 'Saving Product %d' % i
                product_url = product['ImageUrl']
                image_name = 'img_'+str(i)
                urllib.urlretrieve(product_url, os.path.join(images_path,image_name+'.jpg'))
                product_name = product['Name']
                product_desc = product['Description']
                product_price = product['Price']
                jabong_dataframe_imageurl.append(product_url)
                jabong_dataframe_image.append(image_name)
                jabong_dataframe_category.append(product_category)
                jabong_dataframe_desc.append(product_desc)
                jabong_dataframe_price.append(product_price)
                jabong_dataframe_name.append(product_name)
                i+=1
            except:
                print "Unexpected error:", sys.exc_info()[0]
                pass
```

```python
jabong_dataframe['image'] = jabong_dataframe_image
jabong_dataframe['category'] = jabong_dataframe_category
jabong_dataframe['price'] = jabong_dataframe_price
jabong_dataframe['name'] = jabong_dataframe_name
jabong_dataframe['description'] = jabong_dataframe_desc
```

```python
jabong_dataframe.to_csv(csv_path + 'all_jabong.csv')
```
