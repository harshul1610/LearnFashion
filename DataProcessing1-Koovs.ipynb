### Data Preprocessing 1

This code gets all the images from koovs and put them in koovs folder. It also prepares the all_koovs.csv for the tefla ready format. all_koovs is present in koovs folder only. 

```python
import os
import glob
import random
import json
import pandas as pd
import urllib
import sys
```

```python
home_dir = '../Master-Scrapper/json/koovs/'
csv_path = 'koovs/'
images_path = 'koovs/'

try:
    os.mkdir('koovs')
except:
    pass

all_files = glob.glob(home_dir+ '*.json')

i=0
koovs_dataframe = pd.DataFrame()
koovs_dataframe_imageurl = []
koovs_dataframe_category = []
koovs_dataframe_image = []
koovs_dataframe_desc = []
koovs_dataframe_name = []
koovs_dataframe_price = []


for file_ in all_files:
    with open(file_,'r') as f:
        data = f.read()
        json_data = json.loads(data)
        product_category = json_data['category']
        products = json_data['Products']
        for product in products:
            try:
                print 'Saving Product %d' % i
                product_url = product['ImageUrl']
                image_name = 'img_'+str(i)
                urllib.urlretrieve(product_url, os.path.join(images_path,image_name+'.jpg'))
                product_name = product['Name']
                product_desc = product['Description']
                product_price = product['Price']
                koovs_dataframe_imageurl.append(product_url)
                koovs_dataframe_image.append(image_name)
                koovs_dataframe_category.append(product_category)
                koovs_dataframe_desc.append(product_desc)
                koovs_dataframe_price.append(product_price)
                koovs_dataframe_name.append(product_name)
                i+=1
            except:
                print "Unexpected error:", sys.exc_info()[0]
                pass
```

```python
koovs_dataframe['image'] = koovs_dataframe_image
koovs_dataframe['category'] = koovs_dataframe_category
koovs_dataframe['price'] = koovs_dataframe_price
koovs_dataframe['name'] = koovs_dataframe_name
koovs_dataframe['description'] = koovs_dataframe_desc
```

```python
koovs_dataframe.to_csv('all_koovs.csv')
```
